#Funcion Havriliak-Negami

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Para definir una función lo hacemos así:
def HN(omega,tauHN,deps,b,c,einf):
    z=einf + deps*(1+(1j*omega*tauHN)**b)**-c
    return z
# HN es el nombre de la función. el resto son los parámetros que la función usará
#Como la función está definida, podemos usarla desde cualquier parte del notebook
```

```python
#Definimos un rango de frecuencias a usar
freqs = np.logspace(-3,9,140)
```

```python
#Ahora podemos usar la función acá
eps = HN(2*np.pi*freqs,1E-3,1.0,0.5,0.9,3.0)
#La variable "z" en la función, será "eps" en nuestro notebook.
#Notar que los parámetros que le damos a la función están en el mismo orden que los hemos definido: (omega,tauHN,deps,b,c,einf)
```

```python
#Dibujamos los datos obtenidos a través de la función HN

#Parte Real Eps*
plt.semilogx(freqs,eps.real,'ro')
plt.xlabel("f (Hz)")
plt.ylabel("eps'")
plt.show()

#Parte Imaginaria Eps*
plt.semilogx(freqs,-eps.imag,'bo')
plt.xlabel("f (Hz)")
plt.ylabel("eps'")
plt.show()
```

```python
#Notar que ahora podemos llamar nuevamente a la función y calcular otro conjunto de parámetros.
#Podemos asignarlos a otra variable, distinta de eps, por ejemplo eps_repeat

eps_repeat = HN(2*np.pi*freqs,1E-5,2.0,0.5,0.3,3.2)
```

```python
#Dibujamos todos los datos obtenidos a través de la función HN.
#Los puntos rojos y azules corresponden a "eps"
#Las lineas punteadas corresponden a eps_repeat

#Parte Real Eps*
plt.semilogx(freqs,eps.real,'ro')
plt.semilogx(freqs,eps_repeat.real,'k--')
plt.xlabel("f (Hz)")
plt.ylabel("eps'")
plt.show()

#Parte Imaginaria Eps*
plt.semilogx(freqs,-eps.imag,'bo')
plt.semilogx(freqs,-eps_repeat.imag,'k--')
plt.xlabel("f (Hz)")
plt.ylabel("eps'")
plt.show()
```
